```python
import pandas as pd
from sqlalchemy import create_engine
engine = create_engine(!!!!!!! encript credentials with creds file !!!!!!!!!!!!!!!!!!, echo=False)

df1 = pd.read_csv('https://app.periscopedata.com/api/creditninja/chart/csv/9521e869-b643-fac6-8679-e7340cf5ed7e')
df1 = df1.where((pd.notnull(df1)), None)
df1 = df1.add_prefix('so_')
df1.rename(index=str, columns={"so_work_phone": "work_phone",
                               "so_mobile_phone": "mobile_phone", 
                               "so_home_phone": "home_phone",
                               "so_current_loan_leadid" : "current_loan_leadid" }, inplace=True)
df1_head = df1[0:100]
df1_head.to_sql(name='service_objects_features', schema='analytics', con=engine, if_exists = 'replace', index=False)



```
